```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM
```

```python
docs = [
    "india is a great country",
    "india has a rich history",
    "cricket is popular in india",
    "virat kohli is a great batsman",
    "ms dhoni is a legendary captain",
    "sachin tendulkar is god of cricket",
    "we love indian cricket team",
    "india will win the world cup",
    "hard work leads to success",
    "practice makes a person perfect"
]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(docs)

word_index = tokenizer.word_index
vocab_size = len(word_index) + 1

print("Vocabulary Size:", vocab_size)
print(word_index)
```

```python
input_sequences = []

for sentence in docs:
    token_list = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(token_list)):
        n_gram = token_list[:i+1]
        input_sequences.append(n_gram)

print(input_sequences[:10])
```

```python
max_len = max([len(seq) for seq in input_sequences])

input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

print("Max Length:", max_len)
print(input_sequences[:5])
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# One-hot encoding for output
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
model = Sequential()

model.add(Embedding(vocab_size, 50, input_length=max_len-1))
model.add(LSTM(100))
model.add(Dense(vocab_size, activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=100, verbose=1)
```

```python
def predict_next_word(model, tokenizer, text, max_len):
    token_list = tokenizer.texts_to_sequences([text])[0]
    token_list = pad_sequences([token_list], maxlen=max_len-1, padding='pre')

    predicted_id = int(np.argmax(model.predict(token_list), axis=-1)[0])

    for word, index in tokenizer.word_index.items():
        if index == predicted_id:
            return word
```

```python
input_text = "india is a"
print("Input:", input_text)

next_word = predict_next_word(model, tokenizer, input_text, max_len)
print("Next Word:", next_word)
```

```python
model.save("next_word_model.h5")
print("Model saved successfully!")
```
